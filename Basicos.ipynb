### Ejercicio 1: Seleccionar Todos los Productos
**Concepto:** `SELECT` básico

**Descripción:** Selecciona todos los registros de la tabla `Production.Product`.

```sql
SELECT * FROM Production.Product
```

### Ejercicio 2: Filtrar Productos por Precio

**Concepto:** `WHERE` y operadores de comparación

**Descripción:** Selecciona todos los productos cuyo `ListPrice` sea mayor a 100.

```sql
SELECT Name, ListPrice FROM Production.Product WHERE ListPrice > 100;
```

### Ejercicio 3: Ordenar Productos por Nombre

**Concepto:** `ORDER BY`

**Descripción:** Selecciona todos los productos y ordénalos por nombre en orden ascendente.

```sql
SELECT Name, ListPrice FROM Production.Product ORDER BY Name ASC;
```

### Ejercicio 4: Contar el Número de Productos

**Concepto:** `COUNT`

**Descripción:** Cuenta el número total de productos en la tabla `Production.Product`.

```sql
SELECT COUNT(*) AS TotalProducts FROM Production.Product;
```

### Ejercicio 5: Obtener el Precio Máximo de un Producto

**Concepto:** `MAX`

**Descripción:** Obtén el precio máximo de los productos en la tabla `Production.Product`.

```sql
SELECT MAX(ListPrice) AS MaxPrice FROM Production.Product;
```

### Ejercicio 6: Calcular el Precio Promedio de los Productos

**Concepto:** `AVG`

**Descripción:** Calcula el precio promedio de todos los productos en la tabla `Production.Product`.

```sql
SELECT AVG(ListPrice) AS AvgPrice FROM Production.Product;
```

### Ejercicio 7: Seleccionar Productos con Precio entre 50 y 200

**Concepto:** `BETWEEN`

**Descripción:** Selecciona todos los productos cuyo precio esté entre 50 y 200.

```sql
SELECT AVG(ListPrice) AS AvgPrice FROM Production.Product;
```

### Ejercicio 8: Seleccionar Productos cuyo Nombre Empiece con 'A'

**Concepto:** `LIKE`

**Descripción:** Selecciona todos los productos cuyo nombre comience con la letra 'A'.

```sql
SELECT Name FROM Production.Product WHERE Name LIKE 'A%';
```

### Ejercicio 9: Filtrar Productos No Discontinuados

**Concepto:** `WHERE` con `NULL`

**Descripción:** Selecciona todos los productos que no estén discontinuados (`DiscontinuedDate` es `NULL`).

```sql
SELECT Name FROM Production.Product WHERE DiscontinuedDate IS NULL;
```

### Ejercicio 10: Agrupar Productos por Color

**Concepto:** `GROUP BY`

**Descripción:** Agrupa los productos por color y cuenta cuántos productos hay en cada grupo.

```sql
SELECT Color, COUNT(*) AS NumberOfProducts FROM Production.Product GROUP BY Color;
```

### Ejercicio 11: Calcular la Suma de Precios por Color

**Concepto:** `SUM` y `GROUP BY`

**Descripción:** Calcula la suma de los precios por cada color de producto.

```sql
SELECT Color, SUM(ListPrice) AS TotalPrice FROM Production.Product GROUP BY Color;
```

### Ejercicio 12: Filtrar Grupos por Cantidad de Productos

**Concepto:** `HAVING`

**Descripción:** Agrupa los productos por color y selecciona solo los grupos con más de 10 productos.

```sql
SELECT Color, COUNT(*) AS NumberOfProducts FROM Production.Product GROUP BY Color HAVING COUNT(*) > 10;
```

### Ejercicio 13: Seleccionar los 10 Productos Más Caros

**Concepto:** `TOP` y `ORDER BY`

**Descripción:** Selecciona los 10 productos más caros.

```sql
SELECT TOP 10 Name, ListPrice FROM Production.Product ORDER BY ListPrice DESC;
```

### Ejercicio 14: Seleccionar Productos y su Subcategoría

**Concepto:** `JOIN`

**Descripción:** Selecciona todos los productos junto con su nombre de subcategoría

```sql
SELECT p.Name AS ProductName, ps.Name AS SubcategoryName 
FROM Production.Product p JOIN Production.ProductSubcategory ps 
ON p.ProductSubcategoryID = ps.ProductSubcategoryID;
```

### Ejercicio 15: Seleccionar Productos sin Subcategoría

**Concepto:** `LEFT JOIN` y `NULL`

**Descripción:** Selecciona todos los productos que no tienen subcategoría asociada.

```sql
SELECT p.Name AS ProductName 
FROM Production.Product p LEFT JOIN Production.ProductSubcategory ps 
ON p.ProductSubcategoryID = ps.ProductSubcategoryID 
WHERE ps.ProductSubcategoryID IS NULL;
```
